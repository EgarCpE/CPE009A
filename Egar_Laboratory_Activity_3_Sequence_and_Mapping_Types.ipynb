Distance Formula
1. Make a program that would calculate the distance between two points given a list of coordinates. Use the distance
formula.
coorindates_list = [(1,1), (2,3)]

```python
coordinates_list = [(1,1), (2,3)]
x1 = coordinates_list[0][0]
y1 = coordinates_list[0][1]
x2 = coordinates_list[1][0]
y2 = coordinates_list[1][1]

dx = x2 - x1
dy = y2 - y1
distance = (dx**2 + dy**2)**0.5
print(f"The distance is equal to {distance}")
```

Simple Word Filter
2. For a given string input, replace all the words “stupid” with an asterisk * equal to the length of the string. The new
string value should be displayed with the asterisks.

```python
text = "That is a stupid mistake"
target = "stupid"

asterisk = "*" * len(target)

filtered_text = text.replace(target, asterisk)
print(filtered_text)
```

Phonebook
3. Create a simple phonebook program that can read from a list of dictionaries. The program should be able to display a
person’s name, contact, and address based from a user input which is the id of the record.

```python
phonebook = [{"id": 1, "name": "Ja", "contact": "123-456", "address": "25th Japan st."}, {"id": 2, "name": "Rel", "contact": "654-321", "address": "14th Washington st."}]

search_id = int(input("Enter the id of the record: "))

for record in phonebook:

    if record["id"] == search_id:
      print(f"Name: {record['name']}")
      print(f"Contact: {record['contact']}")
      print(f"Address: {record['address']}")
      break
else:
    print("Record not found")
```
