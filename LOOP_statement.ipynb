LOOP statement


Example 1

```python
week = ["Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"]
for x in week:
  print(x)
```

Example 2

```python
week = ["Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"]
for x in week:
  print(x)

  if x == "Thursday":
    break
```

Example 3

```python
week = ["Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"]
for x in week:
  if x == "Thursday":
    break
  print(x)
```

The Range () function

```python
for x in range (6):
  print(x)

for x in range (2,6):
  print(x)
```

```python
for x in range (2,6,2):
  print(x)
```

Nested Loop

```python
adjective = ["red", "big", "tasty"]
fruits = ["apple", "banana", "cherry"]

for x in adjective: #outer loop
  for y in fruits: #inner loop
    print(x,y)
```

example 5 while loop

```python
i = 1
while i < 6:
  print(i)
  i += 1

  #i = i+1 assignment operator
```

The break statement

```python
i = 1
while i < 6:
  print(i)
  if i == 3:
    break
  i += 1
```

The continue statement

```python
i = 0
while i < 6:
  i += 1
  if i == 3:
    continue
  print(i)
```

```python
i = 0
while i < 6:
  i += 1
  if i == 3:
    print(i)
    continue
```

the else statement

```python
i = 1
while i < 6:
  print(i)
  i +=1
else:
  print("i is no longer less than 6")
```
