```python
class Car:
    def __init__(self, name, speed):
        self.name = name
        self.speed = speed

obj1 = Car("Honda", 320)
obj2 = Car("Subaru", 210)

print(obj1.name, obj1.speed)
print(obj2.name, obj2.speed)

def max_speed(self, speed):
    return f"The {self.name} runs at the maximum speed of {speed}km/hr"

print(max_speed(obj1, 320))
print(max_speed(obj2, 210))
```

```python
class Animals:
  def __init__(self,name,habitat,temperature):
    self.name = name
    self.habitat = habitat
    self.temperature = temperature

  def features(self):
    return f"The {self.name} is a {self.temperature}-blooded animal that lives on {self.habitat}."

mammals = Animals("Dog", "Land", "Warm")
birds = Animals("Sparrow", "Air", "Warm")
reptiles = Animals("Snake", "Terrestrial", "Cold")

print(mammals.name)
print(mammals.habitat)

print(birds.name)
print(birds.habitat)

print(reptiles.name)
print(reptiles.habitat)

print(mammals.features())
print(birds.features())
print(reptiles.features())
```

```python
import math

class Shapes:
    def __init__(self, name, sides):
        self.name = name
        self.sides = sides

    def description(self):
        return f"The {self.name} has {self.sides} sides."

    def calculate_area(self, **dimensions):
        name = self.name.lower()

        if name == "triangle":
            return 0.5 * dimensions.get('base', 0) * dimensions.get('height', 0)

        elif name == "circle":
            return math.pi * (dimensions.get('radius', 0) ** 2)

        elif name == "square":
            return dimensions.get('side', 0) ** 2

        elif name == "rectangle":
            return dimensions.get('length', 0) * dimensions.get('width', 0)

        else:
            return "Shape not supported."


shape1 = Shapes("Triangle", 3)
print(shape1.description())
print("Area:", shape1.calculate_area(base=5, height=10))

shape2 = Shapes("Circle", 0)
print(shape2.description())
print("Area:", shape2.calculate_area(radius=7))

shape3 = Shapes("Square", 4)
print(shape3.description())
print("Area:", shape3.calculate_area(side=4))

shape4 = Shapes("Rectangle", 4)
print(shape4.description())
print("Area:", shape4.calculate_area(length=6, width=3))
```
