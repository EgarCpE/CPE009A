```python

i = 1
for i in range(1):
  print('*')
  if i in range(2):
    print('**')
    if i in range(3):
      print('***')
      if i in range(4):
        print('****')
        if i in range(5):
          print('*****')
        if i in range(4):
          print('****')
        if i in range(3):
            print('***')
      if i in range(2):
            print('**')
    if i in range(1):
          print('*')

```
