Develop a python scripts that:

    Calculate total marks and percentage
    Assign grade for each subject based on marks:

90-100: A+

80-89: A

70-79: B+

60-69: B

50-59: C

40-49: D

Below 40: F (Fail)

    Determine overall result (Pass/Fail)



Performance Remarks:


Percentage ≥ 90: "Outstanding"

Percentage ≥ 80: "Excellent"

Percentage ≥ 70: "Very Good"

Percentage ≥ 60: "Good"

Percentage ≥ 40: "Satisfactory"

Below 40: "Needs Improvement"


```python
number_of_subjects = int(input("Enter Number of Subjects: "))
for i in range(number_of_subjects):
    print(f"Subject {i + 1}:")

if number_of_subjects == 2:
  subject_name = str(input("Enter Subject Name: "))
  prelim_grade = float(input("Enter Prelim Grade: "))
  midterm_grade = float(input("Enter Midterm Grade: "))
  final_exam_grade = float(input("Enter Final Grade: "))
  total_grades = prelim_grade + midterm_grade + final_exam_grade



percentage = (total_grades / 300) * 100

print(f"Total Grades for {subject_name}: {total_grades}")
print(f"Percentage: {percentage:.2f}%")

if percentage >= 90:
    print("Outstanding")
    print("Grade: A+")
elif percentage >= 80:
      print("Excellent")
      print("Grade: A")
elif percentage >= 70:
        print("Very Good")
        print("Grade: B+")
elif percentage >= 60:
          print("Good")
          print("Grade: B")
elif percentage >= 50:
          print("Satisfactory")
          print("Grade: C")
elif percentage >= 40:
             print("Satisfactory")
             print("Grade: D")
else:
              print("Needs Improvement")
              print("Grade: F (Fail)")

if percentage >= 40:
    print("Overall Result: Pass")
else:
    print("Overall Result: Fail")
```
